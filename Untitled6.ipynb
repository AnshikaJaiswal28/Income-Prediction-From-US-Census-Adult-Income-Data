```python
# Creating a copy of the DataFrame for modification
data = df.copy()

# Mapping 'sex' column to numerical values
data['sex'] = data['sex'].map({'Male': 0, 'Female': 1})

# Grouping categories in 'marital.status' column
data['marital.status'] = data['marital.status'].replace(['Never-married', 'Divorced', 'Separated', 'Widowed'], 'Single')

# Combining categories in 'marital.status' column
data['marital.status'] = data['marital.status'].replace(['Married-civ-spouse', 'Married-spouse-absent', 'Married-AF-spouse'], 'Married')

# Mapping 'marital.status' column to binary values
data['marital.status'] = data['marital.status'].map({'Married': 1, 'Single': 0})
```

```python
# List of columns to be dropped
columns_to_drop = ['workclass', 'education', 'occupation', 'relationship', 'race', 'native.country']

# Dropping the specified columns
data = data.drop(columns=columns_to_drop, axis=1)
```

```python
#training the model
from sklearn.model_selection import train_test_split


# Define the feature set 'X' by excluding the 'income' column
X = data.drop('income', axis=1)

# Define the target variable 'y' by selecting only the 'income' column
y = data['income']

# Splitting the dataset into training and validation sets
X_train, X_validation, Y_train, Y_validation = train_test_split(X, y, test_size=0.2, random_state=7)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score



# Create a Logistic Regression model
lr_model = LogisticRegression()

# Fit the model using the training data
lr_model.fit(X_train, Y_train)

# Perform 10-fold cross-validation and obtain cross-validation scores
cv_results = cross_val_score(lr_model, X_train, Y_train, cv=10)

# Calculate the mean of the cross-validation scores rounded to four decimal places
lr_mean_score = round(cv_results.mean(), 4)

print("Mean Accuracy Score:", lr_mean_score)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis




# Instantiate the Linear Discriminant Analysis (LDA) model
ldr_model = LinearDiscriminantAnalysis()

# Train the LDA model using the training data
ldr_model.fit(X_train, Y_train)

# Perform 10-fold cross-validation and obtain cross-validation scores
cv_results = cross_val_score(ldr_model, X_train, Y_train, cv=10)

# Calculate the mean of the cross-validation scores rounded to four decimal places
ldr_mean_score = round(cv_results.mean(), 4)

print("Mean Accuracy Score (LDA):", ldr_mean_score)
```

```python
from sklearn.neighbors import KNeighborsClassifier

# Instantiate the K-Nearest Neighbors (KNN) model
knn_model = KNeighborsClassifier()

# Train the KNN model using the training data
knn_model.fit(X_train, Y_train)

# Perform 10-fold cross-validation and obtain cross-validation scores
cv_results = cross_val_score(knn_model, X_train, Y_train, cv=10)

# Calculate the mean of the cross-validation scores rounded to four decimal places
knn_mean_score = round(cv_results.mean(), 4)

print("Mean Accuracy Score (KNN):", knn_mean_score)
```

```python

from sklearn.tree import DecisionTreeClassifier


# Instantiate the Decision Tree model
dt_model = DecisionTreeClassifier()

# Train the Decision Tree model using the training data
dt_model.fit(X_train, Y_train)

# Perform 10-fold cross-validation and obtain cross-validation scores
cv_results = cross_val_score(dt_model, X_train, Y_train, cv=10)

# Calculate the mean of the cross-validation scores rounded to four decimal places
dt_mean_score = round(cv_results.mean(), 4)

print("Mean Accuracy Score (Decision Tree):", dt_mean_score)
```

```python
from sklearn.naive_bayes import GaussianNB


# Instantiate the Gaussian Naive Bayes model
gb_model = GaussianNB()

# Train the model using the training data
gb_model.fit(X_train, Y_train)

# Perform 10-fold cross-validation and obtain cross-validation scores
cv_results = cross_val_score(gb_model, X_train, Y_train, cv=10)

# Calculate the mean of the cross-validation scores rounded to four decimal places
gb_mean_score = round(cv_results.mean(), 4)

print("Mean Accuracy Score (Gaussian Naive Bayes):", gb_mean_score)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis


# Define the parameters for the model as a dictionary
parameters = {'solver': ['svd', 'lsqr', 'eigen']}

# Create a GridSearchCV object with the LDA model and defined parameters
grid_search = GridSearchCV(LinearDiscriminantAnalysis(), parameters, cv=5)

# Fit the GridSearchCV with the training data
grid_search.fit(X_train, Y_train)

# Retrieve the best parameters and best score
best_params = grid_search.best_params_
best_score = grid_search.best_score_

# Access the best model instance
best_ldr_model = grid_search.best_estimator_

print("Best Parameters:", best_params)
print("Best Score:", best_score)
print("Best Linear Discriminant Analysis Model:", best_ldr_model)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report


# Generate predictions for the validation set
y_pred = best_ldr_model.predict(X_validation)

# Calculate the accuracy score
accuracy = accuracy_score(Y_validation, y_pred)

# Generate a confusion matrix
cm = confusion_matrix(Y_validation, y_pred)

# Create a classification report
cr = classification_report(Y_validation, y_pred)
```

```python

from sklearn.metrics import precision_score, recall_score, f1_score

# Assuming 'Y_validation' and 'y_pred' are your true labels and predicted labels, respectively

# Calculate precision score
precision = precision_score(Y_validation, y_pred)

# Calculate recall score
recall = recall_score(Y_validation, y_pred)

# Calculate F1 score
F1_score = f1_score(Y_validation, y_pred)

print("Precision Score:", precision)
print("Recall Score:", recall)
print("F1 Score:", F1_score)
```
